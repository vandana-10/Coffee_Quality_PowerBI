# Splitting The Data into Training and Testing

```python
X=data.drop('Price',axis=1)
Y=data['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3)
```

```python
X.isnull().sum()
```

```python
Y.isnull().sum()
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
LR_model=LinearRegression()
LR_model.fit(X_train,y_train)
```

```python
LR_model.score(X_train,y_train)
```

```python
y_pred=LR_model.predict(X_test)
```

#**Evalution Metrics**

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
from math import sqrt
from sklearn.model_selection import cross_val_score

                                                     # MAE represents avg error between actual and predicted
mean_absolute=mean_absolute_error(y_test,y_pred)     # lower MAE values is better the model predicts
print("Mean Absolute Error is : " , mean_absolute)

r2score=r2_score(y_test,y_pred)                    #R2 represents how well independent variable predicts the out come of dependent variable
print("R2 score is : " ,r2score)                       # Higher R2  value  is better the model predicts


rmse = sqrt(mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error is : " ,rmse)


mean_square=mean_squared_error(y_test,y_pred)
print("Mean Squared Error is : " ,mean_square)


num_folds=5
cross_validation=cross_val_score(LR_model, X, Y,cv=num_folds)
print("CrossValidation is : " ,cross_validation)
```

**Using GridSearchCV to get the best parameters**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV



# Define the hyperparameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'max_features': ['sqrt', 'log2', None]
}

# Create a Random Forest Regressor
rf = RandomForestRegressor()

# Perform GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best hyperparameters:", best_params)
```
